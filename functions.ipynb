```python
n=int(input())
r=int(input())
n_fact=1
for i in range(1,n+1):
    n_fact=n_fact*i
r_fact=1
for i in range(1,r+1):
    r_fact=r_fact*i
n_r_fact=1
for i in range(1,n-r+1):
    n_r_fact=n_r_fact*i
ans=n_fact//(r_fact*n_r_fact)
print(ans)
```

```python
def fact(n):
    n_fact=1
    for i in range(1,n+1):
        n_fact*=i
    return n_fact
```

```python
n=int(input())
r=int(input())
n_fact=fact(n)
r_fact=fact(r)
n_r_fact=fact(n-r)
ans=n_fact//(r_fact*n_r_fact)
print(ans)
```

```python
#function : readable,usable,testing easy,
def isprime(n):
    for i in range(2,n):
        if (n%i==0):
            break
    else:
        return True
    return False
isprime(77)
    
```

```python
def primeFrom2ToN(n):
    for k in range(2,n+1):
        if(isprime(k)):
            print(k)
primeFrom2ToN(100)
```

```python
def ncr(n,r):
    n_fact=fact(n)
    r_fact=fact(r)
    n_r_fact=fact(n-r)
    ans=n_fact//(r_fact*n_r_fact)
    return ans
ncr(5,2)
```

```python
def FToC(s,e,w):
    current=s
    while current<=e:
        fvalue=int((5/9)*(current-32))
        print(current,end="\t")
        print(fvalue)
        current+=w
s=int(input())
e=int(input())
w=int(input())
print(FToC(s,e,w))        
```

```python
#fibonacci number check problem
def checknumber(n):
    if (n==0 or n==1 or n==2):
        return True
    a=0
    b=1
    while (b<n):
        c=a+b
        a=b
        b=c
    if(b==n):
        return True
    return False
n=int(input())
print(checknumber(n))
```

```python
#palindrome number
n=int(input())
temp=n
r=0
while temp!=0:
    r=(r*10)+(temp%10)
    temp//=10
if (n==r):
    print("True")
else:
    print("False")
    
```

```python
def digit(temp):
    d=0
    num=temp
    while(num!=0):
        d+=1
        num=num//10
    return d
```
