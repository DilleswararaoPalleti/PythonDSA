```python
n = int(input())
i = 1
while i <= n:
    j = 1
    start_char = chr(ord('A') + i - 1)
    while j <= n:
        # print jth column
        charP = chr(ord(start_char) + j - 1)
        print(charP, end='')
        j = j + 1
    print()
    i = i + 1
```

```python
n=int(input())
i=1
while i<=n:
    spaces=1
    while( spaces<=n-i):
        print(" ",end=' ')
        spaces+=1
    j=1
    #for increment
    p=i
    while (j<=i):
        print(p,end=' ')
        p+=1
        j+=1
    #for decrement
    k=2*i-2
    while (k>i-1):
        print(k,end=' ')
        k-=1
    print()
    i+=1
```

```python
n=int(input())
i=1
n1=(n+1)/2
n2=n/2
while i<=n1:
    j=1
    while j<=n1-i:
        print(' ',end=' ')
        j+=1
    # for upperpart starts
    k=1
    while k<=(2*i-1):
        print('*',end=' ')
        k+=1
    print()
    i+=1
k=n2
while k>=1:
    #for downspaces
    ds=1
    while ds<=n2-k+1:
        print(' ',end=' ')
        ds+=1
    #for printing starts in down sides
    dst=1
    while dst<(2*k-1):
        print('*',end=' ')
        dst+=1
    print()
    k-=1
```

```python
n=int(input())
i=1
while i<=n:
    #for first half numbers
    j=1
    while (j<=i):
        print(j,end=' ')
        j+=1
    #for spaces
    s=1
    while s<=(2*n-2*i):
        print(' ',end=' ')
        s+=1
    #for numbers
    k=i
    while (k>=1):
        print(k,end=' ')
        k-=1
    print()
    i+=1
```

```python
n=int(input())
if n%2!=0:
    n1=(n+1)/2
    n2=n-n1
    i=1
    while i<=n1:
        # for space in first half
        j=1
        while j<=i-1:
            print(' ',end=' ')
            j+=1
        # for stars
        s=1
        while s<=i:
            print('* ',end=' ')
            s+=1
        print()
        i+=1
    # for down part
    k=1
    while k<=n2:
        #for spaces
        ds=1
        while ds<=(n2-k):
            print(' ',end=' ')
            ds+=1
        #for stars
        dst=1
        while dst<=(n2-k+1):
            print('* ',end=' ')
            dst+=1
        print()
        k+=1
        
        
else:
    print("Not possible")
        
        
    
```
