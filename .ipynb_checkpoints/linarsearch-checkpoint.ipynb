```python
#linear search through the funtions
def linear_search(li,k):
    for i in range(len(li)):
        if li[i]==k:
            return i
    return -1
            
        
```

```python
li=[1,2,3,4,5,6,5,4]
k=9
ind=linear_search(li,k)
print(ind)
```

passing variable through functions


```python
def increment(li):
    li[1]=li[1]+3
    return 
li=[1,2,3,4,5]
increment(li)
print(li)
```

```python
def swap(li):
    for i in range(0,len(li)-1,2):
        li[i],li[i+1]=li[i+1],li[i]
    return li
li=[1,2,3,4,5,6]
swap(li)
```

```python
def reverse(li):
    length=len(li)
    for i in range(length//2):
        li[i],li[length-1-i]=li[length-1-i],li[i]
    print(li)
li=[1,2,3,4,5,6]
reverse(li)
```

```python
def num_list(n):
    return [int(x) for x in str(n)]
```

```python
n=1234567
num_list(n)
```

```python
def digitsum(n):
    while n>=10:
        n=sum(int(d)for d in str(n))
    return n
```

```python
n=12345689
print(digitsum(n))
```

```python
target=7
arr=[1,3,6,2,5,4,3,2,4]
def pairsum(arr,target):
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i]+arr[j]==target:
                print(arr[i],arr[j])
pairsum(arr,7)
    
```

```python
#sort 0 t0 1
arr=[1,0,1,0,2,0]
def sort(arr):
    zeropos=0
    current=0
    while(current<len(arr)):
        if(arr[current]==0):
            arr[current],arr[zeropos]=arr[zeropos],arr[current]
        current+=1
    return arr      
sort(arr)
```
