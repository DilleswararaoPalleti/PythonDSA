```python
n=int(input())
r=int(input())
n_fact=1
for i in range(1,n+1):
    n_fact=n_fact*i
r_fact=1
for i in range(1,r+1):
    r_fact=r_fact*i
n_r_fact=1
for i in range(1,n-r+1):
    n_r_fact=n_r_fact*i
ans=n_fact//(r_fact*n_r_fact)
print(ans)
```

```python
def fact(n):
    n_fact=1
    for i in range(1,n+1):
        n_fact*=i
    return n_fact
```

```python
n=int(input())
r=int(input())
n_fact=fact(n)
r_fact=fact(r)
n_r_fact=fact(n-r)
ans=n_fact//(r_fact*n_r_fact)
print(ans)
```

```python
#function : readable,usable,testing easy,
def isprime(n):
    for i in range(2,n):
        if (n%i==0):
            break
    else:
        return True
    return False
isprime(77)
    
```

```python
def primeFrom2ToN(n):
    for k in range(2,n+1):
        if(isprime(k)):
            print(k)
primeFrom2ToN(100)
```

```python
def ncr(n,r):
    n_fact=fact(n)
    r_fact=fact(r)
    n_r_fact=fact(n-r)
    ans=n_fact//(r_fact*n_r_fact)
    return ans
ncr(5,2)
```

Global Variable and local variables

```python
a=10
def f1():
    v=12
    print(v)
print(a)
# print(v)
f1()
#cant print local variable outside the functions
```

```python
b=13 #global variable
def f():
    global b
    b=10
    return b
print(b)
x=f()
print(b)
print(x)
```

```python
global a3=10
def f3():
    a3=12
    return a3
x=f3()
print(x)
```
