```python
def replace(str,char1,char2):
    newstr=""
    for i in str:
        if (i==char1):
            newstr+=char2
        else:
            newstr+=i
    return newstr
str="fsafsavxttxz"
str=replace(str,"s","a")
str
```

Count the number of vowels, consonants,digits,special characters,spaces

```python

def countInstring(str):
    c,v,d,s=0,0,0,0
    for char in str:
        if((char>='a'and char<='z') or (char>='A' and char<="Z")):
            char=char.lower()
            if(char=='a'or char=='e' or char=='i' or char=='o' or char=='u'):
                v+=1
            else:
                c+=1
        elif(char>='0' and char<='9'):
            d+=1
        else:
            s+=1
    return c,v,d,s
str="My123 Name is Adarsh"
c,v,d,s=countInstring(str)
c,v,d,s
```

```python
str1="abcde"
str2="baedc"

def ispermutation(str1,str2):
    if len(str1)!=len(str2):
        return False
    freq=[0]*256
    for char in str1:
        freq[ord(char)]+=1
    for char in str2:
        freq[ord(char)]-=1
    for i in freq:
        if (i!=0):
            return False
    return True
ispermutation(str1,str2)
```

```python
#remove consecutive duplicates
str="aaabbbaaaccc"
def removedup(str):
    ans=""
    start=0
    n=len(str)
    while start<n:
        uniquechar=start
        nextuniquechar=start+1
        while((nextuniquechar<n) and (str[nextuniquechar]==str[uniquechar])):
            nextuniquechar+=1
        ans=ans+str[uniquechar]
        start=nextuniquechar
    return ans
removedup(str)

    
```

```python
#Compress the string
string="aaabbcddeeeee"
def compress(string):
    ans=""
    count=1
    for i in range(1,len(string)):
        if string[i]==string[i-1]:
            count+=1
        else:
            ans+=string[i-1]+str(count)
            count=1
    ans+=string[-1]+str(count)
    return ans
compress(string)
```
