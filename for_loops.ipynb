```python
# prime number program
start=int(input())
end=int(input())
k=start
while k<end:
    count=0
    for i in range(2,k,1):
        if k%i==0:
            count+=1
            break
    if count==0:
        print(k)
    k+=1
```

```python
from collections import Counter

S = input().strip()

cnt = Counter(S)
mx = max(cnt.values())

ans = ''.join(ch for ch in S if cnt[ch] != mx)
print(ans)
```
