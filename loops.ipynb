```python
n=int(input())
j=2
while j<=n:
    i=2
    flag=True
    while i<j:
        if j%i==0:
            flag=False
        i+=1
    if flag:
        print(j)
    j+=1
```

### Fahrenheit to Celsius

```python
s=int(input())
e=int(input())
w=int(input())
while s<=e:
    cel=(5/9)*(s-32)
    print(f"{int(s)}  {int(cel)}")
    s+=w
```

```python
while True:
    n = int(input("Enter operation (1-5, 6 to exit): "))
    
    if n == 6:
        break
    elif 1 <= n <= 5:
        a = int(input("Enter first number: "))
        b = int(input("Enter second number: "))
        
        if n == 1:
            print(a + b)
        elif n == 2:
            print(a - b)
        elif n == 3:
            print(a * b)
        elif n == 4:
            print(a // b)
        elif n == 5:
            print(a % b)
    else:
        print("Invalid Operation")
```

```python
n=int(input())
temp=n
rev=0
while temp>0:
    rem=temp%10
    rev=rev*10+rem
    temp//=10
if(n==rev):
    print(True)
else:
    print(False)
    
    
```
