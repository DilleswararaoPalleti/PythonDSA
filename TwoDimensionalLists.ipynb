```python
#row wise sum question
test=int(input())
while test>0:
    n,m=map(int,input().split())
    li=[[int(j) for j in input().split()]for ele in range(n)]
    for i in range(n):
        rowsum=0
        for j in range(m):
            rowsum+=li[i][j]
        print(rowsum,end=' ')
    test-=1
```

```python
#largest column sum
def lar_col_sum(li):
    n=len(li)
    m=len(li[0])
    max_sum=0
    index=0
    for j in range(m):
        sum=0
        for i in range(n):
            sum+=li[i][j]
        if(sum>max_sum):
            max_sum=sum
            index=j
    return max_sum,index

li=[[1,2,3,4],[4,5,6,7],[10,11,12,13]]
maxsum,lar_col_index=lar_col_sum(li)
print(maxsum,lar_col_index)
```

```python
#wave print
test=int(input())
while test>0:
    n,m=map(int,input().split())
    li=[[int(ele) for ele in input().split()]for i in range(n)]
    for j in range(m):
        for i in range(n):
            if(j%2==0):
                print(li[i][j],end=' ')
            else:
                k=n-i-1
                print(li[k][j],end=' ')
    test-=1
```

```python
#spiral printing
test=int(input())
while test>0:
    n,m=map(int,input().split())
    li=[[int(ele) for ele in input().split()]for i in range(n)]
    maxcount=int(n*m)
    count=0
    rows,cols=0,0
    rowe,cole=n,m
    while count<maxcount:
        i=cols
        while (i<cole and count<maxcount):
            print(li[rows][i],end=" ")
            count+=1
            i=i+1
        rows+=1
        i=rows
        while (i<rowe and count<maxcount):
            print(li[i][cole-1],end=" ")
            count+=1
            i=i+1
        cole-=1
        i=cole-1
        while (i>=cols and count<maxcount):
            print(li[rowe-1][i],end=" ")
            count+=1
            i-=1
        rowe-=1
        i=rowe-1
        while(i>=rows and count<maxcount):
            print(li[i][cols],end=" ")
            count+=1
            i-=1
        cols+=1
    test-=1
    
```
